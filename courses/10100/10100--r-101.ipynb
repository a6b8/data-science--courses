# Vektoren

```r
v = c( 23.4, 2, 11.2, 10.4, 8.8 )
message( "Vektor:\t\t",  paste( v,collapse=", ") )

r = is( v )
message( "Type:\t\t\t", paste( r, collapse=" " ) )

r = v / 100
message( "Broadcast:\t\t", paste( r, collapse=" " ) )

r = c( 10, 20, 30 ) - c( 1, 2, 3 )
message( "Calulate (1):\t\t", paste( r, collapse=" " ) )

r = 3 * c( 10, 20, 30 )
message( "Calulate (2):\t\t", paste( r, collapse=" " ) )

r = c( 1, 2 ) * c( 1, 2, 3, 4 )
message( "Calulate (3):\t\t", paste( r, collapse=" " ) )

names( v ) = c('CDU', 'SPD', 'FDP', 'Gruene', 'Linke')
message( "Add Index:\t\t" )

v
```
